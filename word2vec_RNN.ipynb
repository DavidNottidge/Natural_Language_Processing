```python
df.columns = ['Review','Sentiment']
```

```python
def text_preprocessor(doc):
    doc = doc.lower().strip()
    try:
        doc = doc.replace('.',". ")
        doc = doc.replace(',',", ")
        doc = doc.replace('!',"! ")
    except:
        pass
    doc = re.sub(r"[^a-z\s]","",doc) #to rmv space and anything not lower
    doc = [d.strip() for d in doc.split() if d not in stopwords]
    doc = " ".join(doc)
    return doc
```

```python
stopwords = nltk.corpus.stopwords.words('English')
```

```python
df['Cleaned_review']=df['Review'].apply(text_preprocessor)
```

Here's a breakdown of the code:

tokenizer.word_index is a dictionary that maps each unique word in the tokenizer's vocabulary to a unique integer index.
len(tokenizer.word_index) returns the number of unique words in the tokenizer's vocabulary.
Adding 1 to this number accounts for an extra index that is reserved for out-of-vocabulary (OOV) words.
token_size is the final result, which represents the total number of unique tokens (i.e., words and OOV tokens) in the tokenizer's vocabulary.
By calculating the token size, you can use it as the input size for a neural network's embedding layer or as the output size for a neural network's final layer, depending on the specific use case.

```python
embedding_matrix = np.zeros((token_size,vec_size))
for word,index in tokenizer.word_index.items():
    if word in word_vec.wv:
        embedding_matrix[index] = word_vec.wv[word]
```
