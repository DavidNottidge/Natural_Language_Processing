#### Text Preprocessing

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
stop_words = nltk.corpus.stopwords.words('english')
stop_words
```

defining a preprocessor function to handle each sentence (document) in the files

```python
def preprocessor(doc):
    doc = doc.lower().strip()
    try:
        doc = doc.replace('.','. ')
        doc = doc.replace(',',', ')
        doc = doc.replace('!','! ')
    except:
        pass
    doc = re.sub(r"[^a-z\s]", "", doc)
    doc = [d.strip() for d in doc.split() if d.strip() not in stop_words]
    doc = ' '.join(doc)
    return doc
```

```python
df['cleaned_review'] = df['Review'].apply(preprocessor)
```

```python
df
```

```python
X = df['cleaned_review']
y = df['Sentiment']
```
